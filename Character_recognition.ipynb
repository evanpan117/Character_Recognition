```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def char2bi(page):
    return [
        0 if pixel == '.' else 1
        for line in page
        for pixel in line
    ]
```

```python
#TRAINING_SET[26][35]
TRAINING_SET = []
TESTING_SET = []
for i in range (26):
    TRAINING_SET.append(char2bi(TRAINING_DATA[i]))
    TESTING_SET.append(char2bi(TEST_DATA[i]))
```

```python
x = np.array(TRAINING_SET)
y = np.identity(26)
b = np.random.rand(26)
w = np.random.rand(26,35)
learning_rate = .01
epoch = 500
error_set = []
```

```python
def threshold(y):
    return 1 if y>0 else 0
```

```python
def learning_algh(epoch_count):
    for epoch in range(epoch_count):    #go through each epoch of trainning
        error_count = 0 
        for neuron_idx in range(26):   #train for each neuron
            for i in range(26):  #go through trainning for each character, i is index for each character
                #prediction    
                y_pred = threshold(np.dot(w[neuron_idx].T, x[i]) + b[neuron_idx])
                #error
                error = (y[neuron_idx][i] - y_pred)            
                #update weights
                for j in range(35):
                    w[neuron_idx][j] += (learning_rate * error * x[i][j])                
                #update bias
                b[neuron_idx] += (learning_rate * error)            
                #count error of predictions
                if y_pred != y[neuron_idx][i]:
                    error_count += 1
        error_set.append(error_count)
```

```python
#training model
learning_algh(epoch)
    
```

```python
x_axil = np.arange(epoch)
plt.plot(x_axil,error_set)
plt.xlabel('epoch')
plt.ylabel('Error')
plt.title('Trainning Progress')
```
